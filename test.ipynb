### Отримую дані з CSV файлу

```python
import pandas as pd

df_csv = pd.read_csv('data.csv')
df_csv.head(2)
```

#### Прибираю зайві стовпчики

```python
new_df_csv = df_csv.drop(['locality_koatuu', 'locality_katottg', 'locality_type', 'locality_name', 'postal_code'],axis=1)
new_df_csv.head(2)
```

```python
# перевіряю та міняю на потрібний формат дані в стовпеці phone
def format_phone_number(phone):
    phone = ''.join(filter(str.isdigit, phone))
    if len(phone) == 9:
        return f"+380{phone}" 
    elif len(phone) == 10 and phone.startswith('0'):
        return f"+38{phone}" 
    else:
        return phone

new_df_csv['phone'] = new_df_csv['phone'].apply(format_phone_number)

# приводжу до відповідного формату стовпець date_of_birth
new_df_csv['date_of_birth'] = pd.to_datetime(new_df_csv['date_of_birth'])

# заміняю в колонці sex True но 'F, False на 'M'
new_df_csv['sex'] = new_df_csv['sex'].replace(True, 'F')
new_df_csv['sex'] = new_df_csv['sex'].replace(False, 'M')

new_df_csv
```

### Отримую дані з API 

```python
import requests

url = "https://raw.githubusercontent.com/lia-diez/fake_json_data/main/fake_data.json"

headers = {
    'x-fake-auth': 'YmlnIGJsYWNrIGNvY2s'
}

response = requests.get(url, headers=headers)

if response.status_code == 200:
    data = response.json()
    df_api = pd.json_normalize(data)
    print(df_api)
else:
    print(f"Помилка: {response.status_code}")
```

```python
df_api.head(3)
```

```python
# Видаляю зайві колонки
new_df_api = df_api.drop(['job', 'district', 'document.type'],axis=1)
new_df_api.head(2)
```

```python
# перевіряю та міняю на потрібний формат дані в стовпці phoneNumber
def format_phone_number(phoneNumber):
    phoneNumber = ''.join(filter(str.isdigit, phoneNumber))
    if len(phoneNumber) == 9:
        return f"+380{phoneNumber}" 
    elif len(phoneNumber) == 10 and phoneNumber.startswith('0'):
        return f"+38{phoneNumber}" 
    else:
        return phoneNumber

new_df_api['phoneNumber'] = new_df_api['phoneNumber'].apply(format_phone_number)

# приводжу до відповідного формату стовпець dateBorn
new_df_api['dateBorn'] = pd.to_datetime(new_df_api['dateBorn'])

# об"єдную document.series та document.number
new_df_api['passport'] = new_df_api.apply(lambda row: f"{row['document.series']} {row['document.number']}" 
                           if row['document.series'] else row['document.number'], axis=1)

# видаляю зайві колонки
new_df_api = new_df_api.drop(['document.series', 'document.number'], axis=1)

new_df_api
```

```python
# змінюю назви колонок
new_df_api = new_df_api.rename(columns={'gender': 'sex', 'firstName': 'name', 'lastName': 'last_name', 'email': 'email_addr','phoneNumber': 'phone', 'dateBorn': 'date_of_birth',})
```

### Oб'єдную два датафрейми

```python
union_df = pd.concat([new_df_csv, new_df_api]).drop_duplicates().reset_index(drop=True)
```
